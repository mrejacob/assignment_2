```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df  = pd.read_csv('D:/Jacob/ADS/Assignments/ADS-Assignment-2-3-main/WA_Fn-UseC_-HR-Employee-Attrition.csv')

#df[['JobRole','Attrition','DistanceFromHome']]

result = df.groupby(["JobRole", "Attrition"],as_index=False)["DistanceFromHome"].count()

print(result)

plt.figure(figsize=(15,10))
sns.barplot(data = result,x="JobRole", y="DistanceFromHome", hue="Attrition")

plt.title("Comparison DistanceFromHome to JobRole & Attrition")
plt.xlabel("JobRole")
plt.ylabel("DistanceFromHome")
```
